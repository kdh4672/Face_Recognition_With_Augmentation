```python
import os
import zipfile
from torchvision.datasets import ImageFolder
import torch
from PIL import Image
import shutil
zip_list = os.listdir('./')
len(zip_list)
```

```python
os.mkdir('./frontal')
os.mkdir('./non_frontal')
# bp = 0
count_f = 0
count_n = 0
for zip_file in zip_list:
#     bp += 1
#     if bp == 3:
#         break
    os.mkdir('./temp')
    os.mkdir('./temp/frontal')
    os.mkdir('./temp/non_frontal')
    one_zip = zipfile.ZipFile(zip_file)
    print(one_zip)
    
    for file in one_zip.namelist():
        if len(file) >= 14:
            if file[0:4] != 'S001':
                continue
            else:
                try:
                    if file[7] == '/':
                        if file[5:7] == 'L1' or file[5:7] == 'L8' or file[5:7] == 'L2':               
                            if file[8:11] == 'E03':
                                continue
                            if 5 <= int(file[13]) <= 9:
                                if int(file[13]) == 7:
                                    one_zip.extract(file,"./temp/frontal")
                                else:
                                    one_zip.extract(file,"./temp/non_frontal")
                    if file[5:8] == 'L22':
                        if file[9:12] == 'E03':
                            continue
                        if 5 <= int(file[14]) <= 9: 
                            if int(file[14]) == 7:
                                one_zip.extract(file,"./temp/frontal")
                            else:
                                one_zip.extract(file,"./temp/non_frontal")
                except:
                    continue
        else:
            continue

    train_dataset = ImageFolder(root='./temp')
    for i in range(len(train_dataset)):
        if train_dataset[i][1] == 0:
            image = train_dataset[i][0]
            image.save('./frontal/{}.png'.format(count_f))
            count_f += 1
        else:
            image = train_dataset[i][0]
            image.save('./non_frontal/{}.png'.format(count_n))
            count_n += 1
    print(count_f/8)
    shutil.rmtree('./temp')
```
